```
from tensorflow.keras.models import load_model
model = load_model("/content/drive/My Drive/Colab Notebooks/NLP mini project/model.h5")
```

```
import pickle
with open('/content/drive/My Drive/Colab Notebooks/NLP mini project/tokenizer.pickle', 'rb') as handle:
    tokenizer = pickle.load(handle)
```

```
for layer in model.layers:
  print(layer.name)
```

```
from tensorflow.keras.models import Model
layer_name = 'global_max_pooling1d_1'
intermediate_layer_model = Model(inputs=model.input,
                                 outputs=model.get_layer(layer_name).output)
```

```
import pandas as pd
df = pd.read_csv("/content/drive/My Drive/Colab Notebooks/NLP mini project/BBC News Train.csv")
```

```
df.columns
```

```
cols = []
for i in range(128):
  cols.append(str(i))
```

```
df_new = pd.DataFrame(columns=cols)
```

```
df_new.info()
```

```
from tensorflow.keras.preprocessing.sequence import pad_sequences
T = 3356
queries = list(df["Text"])
rows = []
for q in queries:
  q = tokenizer.texts_to_sequences([q])
  q = pad_sequences(q, maxlen=T)
  intermediate_output = intermediate_layer_model.predict(q)
  rows.append(intermediate_output[0])
```

```
import numpy as np
data = np.array(rows)
df = pd.DataFrame(data, columns=cols)
df.to_csv('/content/drive/My Drive/Colab Notebooks/NLP mini project/train_embeds.tsv', sep = '\t', index=False, header=False)
```
